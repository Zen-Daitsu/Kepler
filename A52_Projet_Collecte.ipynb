```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
import kaggle
import os
import zipfile
from kaggle.api.kaggle_api_extended import KaggleApi

# Initialize Kaggle API
api = KaggleApi()
api.authenticate()

# Define dataset and download details
dataset = "nasa/kepler-exoplanet-search-results"
download_path ="C:\\Users\\d2s1a\\"  # Directory to save the downloaded files
extract_path = "C:\\Users\\d2s1a\\"  # Directory to extract the CSV
file_name = "cumulative.csv"

# Download the entire dataset (as it may be a zip file)
try:
    api.dataset_download_files(dataset, path=download_path, unzip=False)
    print(f"Successfully downloaded dataset to {'C:\\Users\\d2s1a\\'}")
except Exception as e:
    print(f"Error downloading dataset: {e}")

# Find the downloaded zip file
zip_file_path = None
for file in os.listdir(download_path):
    if file.endswith(".zip"):
        zip_file_path = os.path.join(download_path, file)
        break

if zip_file_path and os.path.exists(zip_file_path):
    # Extract the zip file
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extract(file_name, extract_path)
        print(f"Successfully extracted {file_name} to {extract_path}")
    except Exception as e:
        print(f"Error extracting zip file: {e}")
else:
    print("No zip file found in the download path.")

# Verify the extracted file exists
extracted_file = os.path.join(extract_path, file_name)
if os.path.exists(extracted_file):
    print(f"File {file_name} is present in {extract_path}")
else:
    print(f"File {file_name} was not found in {extract_path}")
```
